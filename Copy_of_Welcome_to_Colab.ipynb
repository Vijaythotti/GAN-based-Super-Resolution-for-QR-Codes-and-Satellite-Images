### **SRGAN Model Architecture**

First, we'll define the Generator and Discriminator networks. The Generator will be responsible for upscaling low-resolution images to high-resolution, and the Discriminator will try to distinguish between real high-resolution images and generated ones.

```
# Generator Block
class ConvBlock(nn.Module):
    def __init__(self, in_channels, out_channels, use_act=True, use_bn=True, **kwargs):
        super().__init__()
        self.conv = nn.Conv2d(in_channels, out_channels, **kwargs)
        self.bn = nn.BatchNorm2d(out_channels) if use_bn else None
        self.act = nn.PReLU() if use_act else None

    def forward(self, x):
        x = self.conv(x)
        if self.bn is not None:
            x = self.bn(x)
        if self.act is not None:
            x = self.act(x)
        return x


class UpsampleBlock(nn.Module):
    def __init__(self, in_channels, scale_factor):
        super().__init__()
        self.conv = ConvBlock(in_channels, in_channels * (scale_factor ** 2), kernel_size=3, stride=1, padding=1)
        self.pixel_shuffle = nn.PixelShuffle(scale_factor) # (C * r^2, H, W) -> (C, H * r, W * r)

    def forward(self, x):
        x = self.conv(x)
        return self.pixel_shuffle(x)


class ResidualBlock(nn.Module):
    def __init__(self, in_channels):
        super().__init__()
        self.block1 = ConvBlock(in_channels, in_channels, kernel_size=3, stride=1, padding=1)
        self.block2 = ConvBlock(in_channels, in_channels, kernel_size=3, stride=1, padding=1, use_act=False)

    def forward(self, x):
        return x + self.block2(self.block1(x))


class Generator(nn.Module):
    def __init__(self, in_channels=3, num_channels=64, num_res_blocks=16):
        super().__init__()
        self.initial = ConvBlock(in_channels, num_channels, kernel_size=9, stride=1, padding=4, use_bn=False)
        self.res_blocks = nn.Sequential(*[ResidualBlock(num_channels) for _ in range(num_res_blocks)])
        self.conv_block = ConvBlock(num_channels, num_channels, kernel_size=3, stride=1, padding=1, use_act=False)
        self.upsample = UpsampleBlock(num_channels, scale_factor=2)
        self.final = nn.Conv2d(num_channels, in_channels, kernel_size=9, stride=1, padding=4)

    def forward(self, x):
        initial = self.initial(x)
        x = self.res_blocks(initial)
        x = self.conv_block(x) + initial
        x = self.upsample(x)
        return torch.tanh(self.final(x))
```

```
# Discriminator Block
class Discriminator(nn.Module):
    def __init__(self, in_channels=3, num_features=[64, 64, 128, 128, 256, 256, 512, 512]):
        super().__init__()
        blocks = []
        for idx, feature in enumerate(num_features):
            blocks.append(ConvBlock(
                in_channels, feature,
                kernel_size=3, stride=1 + idx % 2, padding=1,
                use_act=True, use_bn=False if idx == 0 else True # No BN for first layer
            ))
            in_channels = feature

        self.blocks = nn.Sequential(*blocks)
        self.classifier = nn.Sequential(
            nn.AdaptiveAvgPool2d((6, 6)), # Reduce feature map size
            nn.Flatten(),
            nn.Linear(512 * 6 * 6, 1024),
            nn.LeakyReLU(0.2),
            nn.Linear(1024, 1)
        )

    def forward(self, x):
        x = self.blocks(x)
        return self.classifier(x)
```

```
class ResidualBlock(nn.Module):
    def __init__(self, channels):
        super().__init__()
        self.block = nn.Sequential(
            nn.Conv2d(channels, channels, 3, 1, 1),
            nn.BatchNorm2d(channels),
            nn.PReLU(),
            nn.Conv2d(channels, channels, 3, 1, 1),
            nn.BatchNorm2d(channels)
        )

    def forward(self, x):
        return x + self.block(x)


class Generator(nn.Module):
    def __init__(self, num_res_blocks=16):
        super().__init__()

        self.initial = nn.Sequential(
            nn.Conv2d(3, 64, 9, stride=1, padding=4),
            nn.PReLU()
        )

        self.res_blocks = nn.Sequential(
            *[ResidualBlock(64) for _ in range(num_res_blocks)]
        )

        self.conv = nn.Conv2d(64, 64, 3, 1, 1)
        self.bn = nn.BatchNorm2d(64)

        self.upsample = nn.Sequential(
            nn.Conv2d(64, 256, 3, 1, 1),
            nn.PixelShuffle(2),
            nn.PReLU(),
            nn.Conv2d(64, 256, 3, 1, 1),
            nn.PixelShuffle(2),
            nn.PReLU()
        )

        self.final = nn.Conv2d(64, 3, 9, 1, 4)

    def forward(self, x):
        initial = self.initial(x)
        x = self.res_blocks(initial)
        x = self.bn(self.conv(x)) + initial
        x = self.upsample(x)
        return torch.tanh(self.final(x))
```

```
class Discriminator(nn.Module):
    def __init__(self):
        super().__init__()

        def block(in_c, out_c, stride):
            return nn.Sequential(
                nn.Conv2d(in_c, out_c, 3, stride, 1),
                nn.BatchNorm2d(out_c),
                nn.LeakyReLU(0.2)
            )

        self.model = nn.Sequential(
            nn.Conv2d(3, 64, 3, 1, 1),
            nn.LeakyReLU(0.2),

            block(64, 64, 2),
            block(64, 128, 1),
            block(128, 128, 2),
            block(128, 256, 1),
            block(256, 256, 2),
            block(256, 512, 1),
            block(512, 512, 2),

            nn.AdaptiveAvgPool2d(1),
            nn.Flatten(),
            nn.Linear(512, 1024),
            nn.LeakyReLU(0.2),
            nn.Linear(1024, 1),
            nn.Sigmoid()
        )

    def forward(self, x):
        return self.model(x)
```

```
from fastapi import FastAPI, UploadFile
import torch
import torch.nn as nn
from PIL import Image
import torchvision.transforms as T

# Copied from cell 90caa696 to ensure correct model architecture is used
class ConvBlock(nn.Module):
    def __init__(self, in_channels, out_channels, use_act=True, use_bn=True, **kwargs):
        super().__init__()
        self.conv = nn.Conv2d(in_channels, out_channels, **kwargs)
        self.bn = nn.BatchNorm2d(out_channels) if use_bn else None
        self.act = nn.PReLU() if use_act else None

    def forward(self, x):
        x = self.conv(x)
        if self.bn is not None:
            x = self.bn(x)
        if self.act is not None:
            x = self.act(x)
        return x

# Copied from cell 90caa696 to ensure correct model architecture is used
class UpsampleBlock(nn.Module):
    def __init__(self, in_channels, scale_factor):
        super().__init__()
        self.conv = ConvBlock(in_channels, in_channels * (scale_factor ** 2), kernel_size=3, stride=1, padding=1)
        self.pixel_shuffle = nn.PixelShuffle(scale_factor) # (C * r^2, H, W) -> (C, H * r, W * r)

    def forward(self, x):
        x = self.conv(x)
        return self.pixel_shuffle(x)

# Copied from cell 90caa696 to ensure correct model architecture is used
class ResidualBlock(nn.Module):
    def __init__(self, in_channels):
        super().__init__()
        self.block1 = ConvBlock(in_channels, in_channels, kernel_size=3, stride=1, padding=1)
        self.block2 = ConvBlock(in_channels, in_channels, kernel_size=3, stride=1, padding=1, use_act=False)

    def forward(self, x):
        return x + self.block2(self.block1(x))

# Copied from cell 90caa696 to ensure correct model architecture is used
class Generator(nn.Module):
    def __init__(self, in_channels=3, num_channels=64, num_res_blocks=16):
        super().__init__()
        self.initial = ConvBlock(in_channels, num_channels, kernel_size=9, stride=1, padding=4, use_bn=False)
        self.res_blocks = nn.Sequential(*[ResidualBlock(num_channels) for _ in range(num_res_blocks)])
        self.conv_block = ConvBlock(num_channels, num_channels, kernel_size=3, stride=1, padding=1, use_act=False)
        self.upsample = UpsampleBlock(num_channels, scale_factor=2)
        self.final = nn.Conv2d(num_channels, in_channels, kernel_size=9, stride=1, padding=4)

    def forward(self, x):
        initial = self.initial(x)
        x = self.res_blocks(initial)
        x = self.conv_block(x) + initial
        x = self.upsample(x)
        return torch.tanh(self.final(x))


app = FastAPI()

model = Generator()
model.load_state_dict(torch.load("generator.pth"))
model.eval()

@app.post("/predict/")
async def predict(file: UploadFile):
    image = Image.open(file.file).convert("RGB")
    transform = T.ToTensor()
    img = transform(image).unsqueeze(0)

    with torch.no_grad():
        output = model(img)

    return {"message": "Image processed"}
```
